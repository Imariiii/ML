3. Правда ли, что люди с высоким уровнем стресса имеют более высокое кровяное давление, чем те, кто не подвержен стрессу

````python
high_stress = df[df['Stress Levels'] > df['Stress Levels'].median()]
low_stress = df[df['Stress Levels'] <= df['Stress Levels'].median()]

def parse_blood_pressure(bp_series):
    systolic_values = bp_series.str.split('/').str[0].astype(float)
    return systolic_values

high_stress_bp = parse_blood_pressure(high_stress['Blood Pressure (s/d)']).mean()
low_stress_bp = parse_blood_pressure(low_stress['Blood Pressure (s/d)']).mean()

print(f"Среднее кровяное давление у людей с высоким уровнем стресса: {high_stress_bp:.2f}")
print(f"Среднее кровяное давление у людей с низким уровнем стресса: {low_stress_bp:.2f}")

if (high_stress_bp > low_stress_bp):
    print("Люди с высоким уровнем стресса имеют более высокое кровяное давление")
else:
    print("Люди с высоким уровнем стресса не имеют более высокое кровяное давление")
````

5. Правда ли, что с возрастом женщины имеют свойство оценивать свои когнитивные способности хуже, чем в юности (порог юности - 30 лет, женщина в возрасте - 60 лет+)

````python
women_df = df[df['Gender'] == 'Female']

young_women = women_df[women_df['Age (years)'] <= 30]
old_women = women_df[women_df['Age (years)'] >= 60]

young_women_cognitive_avg = young_women['Cognitive Function'].mean()
old_women_cognitive_avg = old_women['Cognitive Function'].mean()

print(f"Средняя оценка когнитивных способностей у женщин до 30 лет: {young_women_cognitive_avg:.2f}")
print(f"Средняя оценка когнитивных способностей у женщин старше 60 лет: {old_women_cognitive_avg:.2f}")

if old_women_cognitive_avg < young_women_cognitive_avg:
    print("Женщины старше 60 лет оценивают свои когнитивные способности хуже, чем в юности")
else:
    print("Женщины старше 60 лет оценивают свои когнитивные способности лучше или на том же уровне, что и в юности")
````

Влияет, более активные люди имеют тенденцию к более низкому или умеренному весу.

8. Правда ли, что у людей с индексом массы тела больше 75го перцентиля встречаются хронические заболевания чаще, чем у людей с индесом массы тела в переделах от 30 до 60 перцентиля?

Среди людей с индексом массы тела выше 75-го перцентиля (высокий BMI), 296 человек не имеют хронических заболеваний.
Среди людей с BMI в пределах 30-60 перцентиля — 449 человек без хронических заболеваний.
Соответственно, среди людей с более высоким BMI хронические заболевания встречаются реже (учитывая общую численность: 750 человек в группе высокого BMI и 900 в средней группе).

````python
bmi_75_percentile = df['BMI'].quantile(0.75)
bmi_30_percentile = df['BMI'].quantile(0.30)
bmi_60_percentile = df['BMI'].quantile(0.60)

high_bmi = df[df['BMI'] > bmi_75_percentile]
high_bmi_chronic_disease_count = high_bmi[high_bmi['Chronic Diseases'].notnull()].shape[0]
high_bmi_total_count = high_bmi.shape[0]

mid_bmi = df[(df['BMI'] >= bmi_30_percentile) & (df['BMI'] <= bmi_60_percentile)]
mid_bmi_chronic_disease_count = mid_bmi[mid_bmi['Chronic Diseases'].notnull()].shape[0]
mid_bmi_total_count = mid_bmi.shape[0]

high_bmi_percentage = high_bmi_chronic_disease_count / high_bmi_total_count * 100
mid_bmi_percentage = mid_bmi_chronic_disease_count / mid_bmi_total_count * 100

print(f"Процент людей с хроническими заболеваниями при BMI > 75 перцентиля: {high_bmi_percentage:.2f}")
print(f"Процент людей с хроническими заболеваниями при BMI в диапазоне 30-60 перцентилей: {mid_bmi_percentage:.2f}")

if high_bmi_percentage > mid_bmi_percentage:
    print("У людей с BMI выше 75-го перцентиля хронические заболевания встречаются чаще")
else:
    print("У людей с BMI в диапазоне 30-60 перцентилей хронические заболевания встречаются чаще")
````

6. Имеет ли связь зрение и способность слышать звуки на высоких частотах? Ответьте на вопрос - какая зависимость между ними и предположите почему зависимость именно такая?

Между данными показателями линейная зависимость


````python
plt.figure(figsize=(10, 6))
sns.scatterplot(x=df['Vision Sharpness'], y=df['Hearing Ability (dB)'], hue=df['Age (years)'], palette="magma")

bins = np.linspace(df['Vision Sharpness'].min(), df['Vision Sharpness'].max(), 30)

df['Vision Bin'] = pd.cut(df['Vision Sharpness'], bins)

mean_values = df.groupby('Vision Bin', observed=False).agg({
    'Vision Sharpness': 'mean',
    'Hearing Ability (dB)': 'mean'
}).reset_index()

first_x = mean_values['Vision Sharpness'].iloc[0]
last_x = mean_values['Vision Sharpness'].iloc[-1]

first_y = mean_values['Hearing Ability (dB)'].iloc[0]
last_y = mean_values['Hearing Ability (dB)'].iloc[-1]

plt.plot([first_x, last_x], [first_y, last_y], color='blue', linestyle='-', label='Линейная регрессия')

plt.title('Связь между зрением и способностью слышать высокие частоты')
plt.xlabel('Зрение (Vision Sharpness)')
plt.ylabel('Слух на высоких частотах (dB)')
plt.legend()

plt.show()
````
